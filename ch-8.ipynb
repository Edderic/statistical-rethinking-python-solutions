## Easy

### 8E1

Which of the following is a requirement of the simple Metropolis algorithm?

(a) The parameters must be discrete.
      
    No. The algorithm also works for continuous parameters.

(b) The likelihood function must be Gaussian.
    
    No. MCMC could estimate parameters that have non-Gaussian distributions, unlike quadratic approximations that were used in earlier chapters.

(c) The proposal distribution must be symmetric.
    
    Yes, the distribution has to be symmetric.

### 8E2

Gibbs sampling is more efficient than the Metropolis algorithm. How does it achieve this extra efficiency? Are there any limitations to the Gibbs sampling strategy?

#### Ans

It gets its extra efficiency by making use of conjugate pairs, which have analytical solutions and therefore could be quick to compute. 

Limitations:

1. Maybe a conjugate prior does not make sense -- Gibbs sampling forces us to use priors that speed up computation, but might not make sense for the problem at hand.

2. Calculations are very slow when there are hundreds or thousands of parameters.

### 8E3

Which sort of parameters can Hamiltonian Monte Carlo not handle? Can you explain why?

#### Ans

HMC can only handle continuous parameters (not discrete). This is because it "glides" through a surface in order to work / involves gradient calculations.

### 8E4

Explain the difference between the effective number of  samples, `n_eff` as calculated by Stan, and the actual number of samples.

#### Ans

`n_eff` is an estimate of the number of independent samples. It is generally smaller than the actual number of samples (which will have autocorrelation). Low $\frac{n_{eff}}{\text{# actual samples}}$ shows that the chains are inefficient, but possibly still okay. A high ratio is a good sign.

### 8E5

Which value should `Rhat` approach, when a chain is sampling the posterior distribution correctly?

#### Ans

It should approach 1.0

### 8E6

Sketch a good trace plot for a Markov chain, one that is effectively sampling from the posterior distribution. What is good about its shape? Then sketch a trace plot for a malfunctioning Markov chain. What about its shape indicates malfunction?

#### Ans

![Good and bad chain](img/good-bad-chain.jpg)

A good chain has _stationarity_ and _good mixing_. _Stationarity_ means that the samples stick around a very stable central tendency. _Good mixing_, on the other hand, means that there isn't a high degree of correlation between successive samples.

### 8M1

Re-estimate the terrain ruggedness model from the chapter, but now using a uniform prior and an exponential prior for the standard deviation, `sigma`. The uniform prior should be `dunif(0,10)` and the exponential should be `dexp(1)`. Do the different priors have any detectible influence on the posterior distribution?

```python
rugged_df = pd.read_csv('./Data/rugged.csv', sep=';')
rugged_df['log_gdp'] = np.log(rugged_df['rgdppc_2000'])
dd = rugged_df[rugged_df.notnull()]
dd
```

```python
with pm.Model() as model_8m1_unif_0_10:
    alpha = pm.Normal('alpha', 0, 100)
    bR = pm.Normal('bR', 0, 10)
    bA = pm.Normal('bA', 0, 10)
    bAR = pm.Normal('bAR', 0, 10)
    mu = alpha + bR * dd['rugged'] + \
        bA * dd['cont_africa'] + \
        bAR * dd['rugged'] * dd['cont_africa']
    sigma = pm.Uniform('sigma', 0, 10)
    
    obs = pm.Normal('observed', mu, sigma, observed=dd['log_gdp'])
    trace_8m1_unif_0_10 = pm.sample(2000)
```

```python
pm.traceplot(trace_8m1_unif_0_10)
```

```python
with pm.Model() as model_8m1_exp_1:
    alpha = pm.Normal('alpha', 0, 100)
    bR = pm.Normal('bR', 0, 10)
    bA = pm.Normal('bA', 0, 10)
    bAR = pm.Normal('bAR', 0, 10)
    mu = alpha + bR * dd['rugged'] + \
        bA * dd['cont_africa'] + \
        bAR * dd['rugged'] * dd['cont_africa']
    sigma = pm.Exponential('sigma', 1)
    
    obs = pm.Normal('observed', mu, sigma, observed=dd['log_gdp'])
    trace_8m1_exp_1 = pm.sample(2000)
```

### 8H2

Recall the divorce rate example from Chapter 5. Repeat that analysis, using `map2stan` this time, fitting models `m5.1`, `m5.2`, and `m5.3`. Use `compare` to compare the models on the basis of WAIC. Explain the results.

```python
waffle_df = pd.read_csv('./Data/WaffleDivorce.csv', sep=';')

waffle_df['std_MedianAgeMarriage'] = (waffle_df['MedianAgeMarriage'] - waffle_df['MedianAgeMarriage'].mean()) / waffle_df['MedianAgeMarriage'].std()
waffle_df['std_Marriage'] = (waffle_df['Marriage'] - waffle_df['Marriage'].mean()) / waffle_df['Marriage'].std()

waffle_df
```

```python
with pm.Model() as m_5_1:
    alpha = pm.Normal('alpha', 10, 10)
    bA = pm.Normal('bA', 0, 1)
    mu = alpha + bA * waffle_df['std_MedianAgeMarriage']
    sigma = pm.Uniform('sigma', 0, 10)
    divorce = pm.Normal('divorce', mu, sigma, observed=waffle_df['Divorce'])
    
    trace_5_1 = pm.sample(2000)
```

```python
with pm.Model() as m_5_2:
    alpha = pm.Normal('alpha', 10, 10)
    bR = pm.Normal('bR', 0, 1)
    mu = alpha + bR * waffle_df['std_Marriage']
    sigma = pm.Uniform('sigma', 0, 10)
    divorce = pm.Normal('divorce', mu, sigma, observed=waffle_df['Divorce'])
    
    trace_5_2 = pm.sample(2000)
```

```python
with pm.Model() as m_5_3:
    alpha = pm.Normal('alpha', 10, 10)
    bR = pm.Normal('bR', 0, 1)
    bA = pm.Normal('bA', 0, 1)
    mu = alpha + bR * waffle_df['std_Marriage'] + bA * waffle_df['std_MedianAgeMarriage']
    sigma = pm.Uniform('sigma', 0, 10)
    divorce = pm.Normal('divorce', mu, sigma, observed=waffle_df['Divorce'])
    
    trace_5_3 = pm.sample(2000)
```

```python
pm.compare(traces = [trace_5_1, trace_5_2, trace_5_3], models = [m_5_1, m_5_2, m_5_3])
```

#### Ans

The worst model is the one with only marriage rate as sole predictor. Adding median age of marriage to the mix significantly improves WAIC. However, there is a slight improvement in WAIC after removing marriage rate from the second model. This suggests that the information given by marriage rate is redundant when marriage rate is also considered in the same model.

Go back to the leg length example in Chapter 5. Here is the code again, which simulates height and leg lengths for 100 imagined individuals:

    N <- 100                                 # number of individuals
    height <- rnorm(N,10,2)                  # sim total height of each
    leg_prop <- runif(N,0.4,0.5)             # leg as proportion of height
    leg_left <- leg_prop*height +            # sim left leg as proportion + error
        rnorm(N, 0, 0.2)
    leg_right <- leg_prop*height +            # sim right leg as proportion + error
        rnorm(N, 0, 0.2) 
                                             # combine into dataframe
    d <- data.frame(height, leg_left, leg_right)
