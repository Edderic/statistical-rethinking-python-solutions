# Easy

## 6E1

State the three motivating criteria that define information entropy. Try to express each in your own words.

### Ans

1. The measure of uncertainty should be continuous (i.e. arbitrarily small changes in probabilities should not make big jumps in this measure).

2. The measure of uncertainty should increase as the number of possible events increases (i.e. having more "options" increases uncertainty). For example, building an accurate binary classifier is generally easier than building an accurate multiclass classifier.

3. The measure of uncertainty should be additive (i.e. to measure total uncertainty, uncertainty about one feature can be added to uncertainty about another feature).

## 6E2

Suppose a coin is weighted such that, when it is tossed and lands on a table, it comes up heads 70% of the time. What is the entropy of this coin?

### Ans

```python
-(0.7 * np.log(0.7) + 0.3 * np.log(0.3))
```

## 6E3

Suppose a four-sided die is loaded such that, when tossed onto a table, it shows "1" 20%, "2" 25%, "3" 25%, and "4" 30% of the time. What is the entropy of this die?

### Ans

```python
-sum([i * np.log(i) for i in [0.2, 0.25, 0.25, 0.30]])
```

## 6E4

Suppose another four-sided die is loaded such that it never shows "4". The other three sides show equally often. What is the entropy of this die?

### Ans

We skip calculating $p_4\text{log}(p_4)$ because we know that it $p_4$ is never going to happen.

```python
-sum([i * np.log(i) for i in [1.0/3, 1.0/3, 1.0/3]])
```

## 6M1

Write down and compare the definitions of AIC, DIC, and WAIC. Which of these criteria is most general? Which assumptions are required to transform a more general criterion into a less general one?

### Ans

$$
\begin{equation}
\begin{aligned}
    \text{AIC} &= D_{train} + 2p
\end{aligned}
\end{equation}
$$

$D_{train}$ is deviance $-2 \sum_i \text{log} q_i $ on the training set, where $q_i$ is the likelihood of observing $y_i$, based on the model. $p$ is the number of parameters.

$$
\begin{equation}
\begin{aligned}
    \text{DIC} &= \bar{D} + (\bar{D} - \hat{D}) \\
    &= \bar{D} + p_D
\end{aligned}
\end{equation}
$$


$\bar{D}$ is the deviance distribution of the sample. $\hat{D}$ is the average deviance of the posterior distribution.

$\text{Pr(}y_i\text{)}$ is the average likelihood of observation $i$ in the training sample.

log-pointwise-predictive-density:

$$
\begin{equation}
\begin{aligned}
    \text{lppd} &= \sum_{i=1}^N \text{log Pr} (y_i) \\
    p_{\text{WAIC}} &= \sum_{i=1}^N V(y_i) \\
    \text{WAIC} &= -2(\text{lppd} - p_{\text{WAIC}})
\end{aligned}
\end{equation}
$$

To calculate `lppd`, we use $n$ samples of the posterior distribution, take the log-likelihood of observation $i$ for each sample of the posterior, giving us $n$ items per $i$-th observation. Then, we take the average of the likelihood per observation, thereby giving us 1 item per $i$-th observation (via the log scale, to be precise with our estimate). Summing that up gives us `lppd`.

$P_{\text{WAIC}}$ is calculated by taking the variance of the log-likelihoods per observation, and then summing it up.

The most general is WAIC (Widely Applicable Information Criteria) and does not require a multivariate Gaussian posterior, unlike DIC. The least general is AIC, which assumes not only a multivariate Gaussian posterior, but also assumes non-informative (i.e. flat priors) and that the sample size is much greater than the number of parameters.

## 6M2

Explain the difference between model _selection_ and model _averaging_. What information is lost under model selection? What information is lost under model averaging?

### Ans

_Selection_ implies that we would only use one model instead of keeping others around. We lose uncertainty of other models since we are throwing them out. _Averaging_, on the other hand, implies that we are keeping a bunch of models to create an ensemble, which should keep uncertainty about our predictions. Little information is lost under model averaging, relative to model selection.

## 6M3

When comparing models with an information criterion, why must all models be fit to exactly the same observations? What would happen to the information criterion values, if the models were fit to different numbers of observations? Perform some experiments, if you are not sure.

### Ans

All models must be fit to exactly the same observations because we are trying to have an apples-to-apples comparison of models. If we were to fit models with different data, then it would be hard to say whether or not one model is better than another. For example, let's say we've observed that model $A$ scores better in WAIC than model $B$. Was it because of the structure of model A? Or was it because model A got lucky and was trained on data that was easier to predict? If we fit the models using the same observations, then we wouldn't run into this issue.

If the models were fit to different numbers of observations, it would make it more cognitively difficult for us to compare estimates, since WAIC, for example, scales with sample size.

## 6M4

What happens to the effective number of parameters, as measured by DIC or WAIC, as a prior becomes more concentrated? Why? Perform some experiments, if you are not sure.

As a prior becomes more concentrated, the effective number of parameters should go down. For example, $p_{\text{WAIC}}$ considers the variance of the log-likelihoods. If we were to concentrate the prior in one area, that would decrease the variance, and therefore the "effective number of parameters."

## 6M5

Provide an informal explanation of why informative priors reduce overfitting.

### Ans

Informative priors reduce overfitting by constricting the possibilities to a small range of values, thereby weakening the model's ability to learn from data (and the noise that comes with that particular sample).

## 6M6

Provide an informal explanation of why overly informative priors result in underfitting

### Ans

Overly informative priors means that the model would be putting lots of emphasis on a a small range of parameter values. So even though we collect data that might show a trend, the model might not learn that trend because of too-strong priors.

```python
howell_df = pd.read_csv('./Data/Howell1.csv', sep=';')
howell_df
```

```python
train_howell, test_howell = train_test_split(howell_df, train_size=0.5)
```

```python
train_howell['age.std'] = (train_howell['age'] - train_howell['age'].mean()) / \
                            train_howell['age'].std()

train_howell['height.std'] = (train_howell['height'] - train_howell['height'].mean()) / \
                            train_howell['height'].std()
    
test_howell['age.std'] = (test_howell['age'] - test_howell['age'].mean()) / \
                            test_howell['age'].std()

test_howell['height.std'] = (test_howell['height'] - test_howell['height'].mean()) / \
                            test_howell['height'].std()
```

#### Train Models

```python
with pm.Model() as m1_6h1:
    alpha = pm.Normal('a', 0,10)
    beta_1 = pm.Normal('b_1', 0, 10)
    mu = pm.Deterministic(
        'mu',
        alpha + beta_1 * train_howell['age.std']
    )
    sigma = pm.Uniform('sigma', 0, 10)
    height_hat = pm.Normal('height_hat', mu, sigma,
                          observed=train_howell['height.std'])
    trace_m1_6h1 = pm.sample(1000, tune=1000)
    
with pm.Model() as m2_6h1:
    alpha = pm.Normal('a', 0,10)
    beta_1 = pm.Normal('b_1', 0, 10)
    beta_2 = pm.Normal('b_2', 0, 10)
    mu = pm.Deterministic(
        'mu',
        alpha + beta_1 * train_howell['age.std'] \
        + beta_2 * train_howell['age.std']**2
    )
    sigma = pm.Uniform('sigma', 0, 10)
    height_hat = pm.Normal('height_hat', mu, sigma,
                          observed=train_howell['height.std'])
    trace_m2_6h1 = pm.sample(1000, tune=1000)

with pm.Model() as m3_6h1:
    alpha = pm.Normal('a', 0,10)
    beta_1 = pm.Normal('b_1', 0, 10)
    beta_2 = pm.Normal('b_2', 0, 10)
    beta_3 = pm.Normal('b_3', 0, 10)
    mu = pm.Deterministic(
        'mu',
        alpha + beta_1 * train_howell['age.std'] \
        + beta_2 * train_howell['age.std']**2 \
        + beta_3 * train_howell['age.std']**3
    )
    sigma = pm.Uniform('sigma', 0, 10)
    height_hat = pm.Normal('height_hat', mu, sigma,
                          observed=train_howell['height.std'])
    trace_m3_6h1 = pm.sample(1000, tune=1000)

with pm.Model() as m4_6h1:
    alpha = pm.Normal('a', 0,10)
    beta_1 = pm.Normal('b_1', 0, 10)
    beta_2 = pm.Normal('b_2', 0, 10)
    beta_3 = pm.Normal('b_3', 0, 10)
    beta_4 = pm.Normal('b_4', 0, 10)
    mu = pm.Deterministic(
        'mu',
        alpha + beta_1 * train_howell['age.std'] \
        + beta_2 * train_howell['age.std']**2 \
        + beta_3 * train_howell['age.std']**3 \
        + beta_4 * train_howell['age.std']**4
    )
    sigma = pm.Uniform('sigma', 0, 10)
    height_hat = pm.Normal('height_hat', mu, sigma,
                          observed=train_howell['height.std'])
    trace_m4_6h1 = pm.sample(1000, tune=1000)

with pm.Model() as m5_6h1:
    alpha = pm.Normal('a', 0,10)
    beta_1 = pm.Normal('b_1', 0, 10)
    beta_2 = pm.Normal('b_2', 0, 10)
    beta_3 = pm.Normal('b_3', 0, 10)
    beta_4 = pm.Normal('b_4', 0, 10)
    beta_5 = pm.Normal('b_5', 0, 10)
    mu = pm.Deterministic(
        'mu',
        alpha + beta_1 * train_howell['age.std'] \
        + beta_2 * train_howell['age.std']**2 \
        + beta_3 * train_howell['age.std']**3 \
        + beta_4 * train_howell['age.std']**4 \
        + beta_5 * train_howell['age.std']**5
    )
    sigma = pm.Uniform('sigma', 0, 10)
    height_hat = pm.Normal('height_hat', mu, sigma,
                          observed=train_howell['height.std'])
    trace_m5_6h1 = pm.sample(1000, tune=1000)
    
with pm.Model() as m6_6h1:
    alpha = pm.Normal('a', 0,10)
    beta_1 = pm.Normal('b_1', 0, 10)
    beta_2 = pm.Normal('b_2', 0, 10)
    beta_3 = pm.Normal('b_3', 0, 10)
    beta_4 = pm.Normal('b_4', 0, 10)
    beta_5 = pm.Normal('b_5', 0, 10)
    beta_6 = pm.Normal('b_6', 0, 10)
    mu = pm.Deterministic(
        'mu',
        alpha + beta_1 * train_howell['age.std'] \
        + beta_2 * train_howell['age.std']**2 \
        + beta_3 * train_howell['age.std']**3 \
        + beta_4 * train_howell['age.std']**4 \
        + beta_5 * train_howell['age.std']**5 \
        + beta_6 * train_howell['age.std']**6
    )
    sigma = pm.Uniform('sigma', 0, 10)
    height_hat = pm.Normal('height_hat', mu, sigma,
                          observed=train_howell['height.std'])
    trace_m6_6h1 = pm.sample(1000, tune=1000)
```

```python
compare_df = pm.compare(
    [trace_m1_6h1, trace_m2_6h1, trace_m3_6h1, trace_m4_6h1, trace_m5_6h1, trace_m6_6h1],
    [m1_6h1, m2_6h1, m3_6h1, m4_6h1, m5_6h1, m6_6h1], 
    method='pseudo-BMA')
```
