For each of the causal relationships below, name a hypothetical third variable that would lead to an interaction effect.

##### Ans
Bread dough rises because of yeast AND temperature. Without enough temperature, even with yeast present, bread dough would not rise.

(2) A car will go faster when it has more cylinders or when it has a better fuel injector.

(3) Most people acquire their political beliefs from their parents, unless they get them instead from their friends.

#### Ans

### 7E3

For each of the explanations in 7E2, write a linear model that expresses the stated relationship.

#### Ans

(1) Let $c_i$ be the caramelization level,  $h_i$ be the heat level, $d_i$ be the dryness of the onions at observation $i$

$$
\begin{equation}
\begin{aligned}
  c_i &\sim \text{Normal}(\mu_i, \sigma)\\
  \mu_i &=\alpha + \beta_h h_i + \beta_d d_i + \beta_{hd} h_i d_i
\end{aligned}
\end{equation}
$$

(2) Let $s_i$ be the speed level, $c_i$ be the number of cylinders, $p_i$ be the fuel projectors at observation $i$

$$
\begin{equation}
\begin{aligned}
  s_i &\sim \text{Normal}(\mu_i, \sigma)\\
  \mu_i &= \alpha + \beta_c c_i + \beta_p p_i & \text{assuming no interaction}
\end{aligned}
\end{equation}
$$

(3) Let $p_i$ be 1 if child's belief came from parent, $f_i$ be the event of whether or not belief came from the $i$-th child's friend.

$$
\begin{equation}
\begin{aligned}
  p_i &\sim \text{Normal}(\mu_i, \sigma)\\
  \mu_i &= \alpha + \beta_f f_i & \beta_f \text{ is probably going to be negative}
\end{aligned}
\end{equation}
$$

(4) Let $k_i$ represent the animal's intelligence, $s_i$ be socialness, $m_i$ be the presence of manipulative appendages

$$
\begin{equation}
\begin{aligned}
  k_i &\sim \text{Normal}(\mu_i, \sigma)\\
  \mu_i &= \alpha + \beta_s s_i + \beta_m m_i + \beta_{sm} s_i m_i
\end{aligned}
\end{equation}
$$

Recall the tulips example from the chapter. Suppose another set of treatments adjusted the temperature in the greenhouse over the two levels: cold and hot. the data in the chapter were collected at the cold temperature. you find none of the plants grown under the hot temperature developed any blooms at all, regardless of the water and shade levels. Can you explain this result in terms of interactions between water, shade, and temperature?

#### Ans

$$
\begin{equation}
\begin{aligned}
  b_i &\sim \text{Normal}(\mu, \sigma)\\
  \mu_i &= w_i \beta_w + \beta_s s_i + \beta_{ws}w_i s_i + \beta_{wt} w_i t_i + \beta_{st} s_i t_i + \beta_{wst} w_i s_i t_i \\
  &= w_i \beta_w + \beta_s s_i + \beta_{ws}w_i s_i + ( \beta_{wt} w_i + \beta_{st} s_i + \beta_{wst} w_i s_i )t_i \\
\end{aligned}
\end{equation}
$$


Let $t_i$ represent temperature. When temperature is low (or close to zero), then we get the original equation (as if the temperature variable didn't exist). However, when temperature rises, our estimates for $\mu$ approach to 0, thanks to (probably negative) weights associated to $t_i$.

### 7M3

In parts of North America, ravens depend upon wolves for their food. This is because ravens are carnivorous but cannot usually kill or open carcasses of prey. Wolves however can and do kill and tear open animals, and they tolerate ravens co-feeding at their kills. This species relationship is generally described as a "species interaction." Can you invent a hypothetical set of data on raven population size in which this relationship would manifest as a statistical interaction? Do you think the biological interaction could be linear? Why or why not?

#### Ans

Let $R_i$ be the number of ravens found in the sample. Let $R_{i+1}$ be the number of ravens found the year after. Let $W_i$ be the number of wolves in $i$-th observation.

$$
\begin{equation}
\begin{aligned}
   R_{i+1} &\sim \text{Normal}(\mu_i, \sigma)\\
   \mu_i &\sim W_i(\beta_W + \beta_{WR}R_i) + \beta_R R_i
\end{aligned}
\end{equation}
$$

When the number of wolves is 0, then the predicted number of ravens next year only depends on the number of ravens this year (times its coefficient). However, initially, when the number of wolves increases, we expect that the number of ravens should increase by some amount. Then we might get a linear increase for a while. After a while, however,the number of wolves might be too high, it's possible that the food reserves might be depleted and follow more of an S-curve.

### 7H1

Return to the `data(tulips)` example in the chapter. Now include the `bed` variable as a predictor in the interaction model. Don't interact `bed` with the other predictors; just include it as a main effect. Note that `bed` is categorical. So to use it properly, you will need to either construct dummy variables or rather an index variable, as explained in Chapter 6.

#### Ans

```python
tulips_df = pd.read_csv('./Data/tulips.csv', sep=';')
tulips_df
```

```python
tulips_df['bed'].unique()
```

```python
tulips_df['bed_index'] = tulips_df['bed'].replace({'a': 0, 'b': 1, 'c': 2})
tulips_df
```

```python
tulips_df['bed.std'] = (tulips_df['bed_index'] - tulips_df['bed_index'].mean()) / tulips_df['bed_index'].std()
tulips_df['water.std'] = (tulips_df['water'] - tulips_df['water'].mean()) / tulips_df['water'].std()
tulips_df['shade.std'] = (tulips_df['shade'] - tulips_df['shade'].mean()) / tulips_df['shade'].std()
tulips_df['blooms.std'] = (tulips_df['blooms'] - tulips_df['blooms'].mean()) / tulips_df['blooms'].std()

tulips_df
```

```python
with pm.Model() as m_7h1:
    a = pm.Normal('a', 0, 1)
    b_bed = pm.Normal('b_bed', 0, 1)
    b_water = pm.Normal('b_water', 0, 1)
    b_shade = pm.Normal('b_shade', 0, 1)
    mu = a + b_bed * tulips_df['bed.std'] \
        + b_water * tulips_df['water.std'] \
        + b_shade * tulips_df['shade.std']
    sigma = pm.Uniform('sigma', 0,5)
    blooms_hat = pm.Normal('blooms_hat', mu, sigma, observed=tulips_df['blooms.std'])
    trace_m_7h1 = pm.sample(2000)
```

### 7H2

Use WAIC to compare the model from *7H1* to a model that omits `bed`. What do you infer from this comparison? Can you reconcile the WAIC results with the posterior distribution of the `bed` coefficients?

#### Ans

```python
with pm.Model() as m_7h1_without_bed:
    a = pm.Normal('a', 0, 1)
    b_water = pm.Normal('b_water', 0, 1)
    b_shade = pm.Normal('b_shade', 0, 1)
    mu = a + b_water * tulips_df['water.std'] \
        + b_shade * tulips_df['shade.std']
    sigma = pm.Uniform('sigma', 0,5)
    blooms_hat = pm.Normal('blooms_hat', mu, sigma, observed=tulips_df['blooms.std'])
    trace_m_7h1_without_bed = pm.sample(2000)
```

```python
compare_df = pm.compare(
    traces=[trace_m_7h1, trace_m_7h1_without_bed], 
    models=[m_7h1, m_7h1_without_bed]
)

compare_df.index = ['m_7h1', 'm_7h1_without_bed']
compare_df
```

```python
pm.compareplot(compare_df)
```

```python
pm.forestplot(trace_m_7h1)
```

```python
pm.forestplot(trace_m_7h1_without_bed)
```

Adding the `bed` predictor is an improvement over the model without it. Comparing the posterior distributions of the two models, it looks like they are essentially the same (the common parameters didn't really change at all), except that the one with the `bed` predictor has a posterior distribution for `bed` that is mostly in the positive yet smaller effect than water and shade.

### 7H3

Consider again the `data(rugged)` data on economic development and terrain ruggedness, examined in this chapter. one of the African countries in that example, Seychelles, is far outside the cloud of other nations, being a rare country with both relatively high GDP and high ruggedness. Seychelles is also unusual, in that it is a group of islands far from the coast of mainland Africa, and its main economic activity is tourism.

One might suspect that this one nation is exerting a strong influence on the conclusions. In this problem, I want you to drop Seychelles from the data and re-evaluate the hypothesis that the relationship of African economies with ruggedness is different from that on other continents.

#### (a) 

Begin by using `map` to fit just the interaction model:

$$
\begin{equation}
\begin{aligned}
y_i &\sim \text{Normal}(\mu_i, \sigma) \\
\mu_i &= \alpha + \beta_A A_i + \beta_R R_i + \beta_{AR} A_i R_i \\
\end{aligned}
\end{equation}
$$

where $y$ is log GDP per capita in the year 2000 (log of `rgdppc_2000`); $A$ is `cont_africa`, the dummy variable for being an African nation; and $R$ is the variable rugged. Choose your own priors. Compare the inference from this model fit to the data without Seychelles to the same model fit to the full data. Does it still seem like the effect of ruggedness depends upon continent? how much has the expected relationship changed?


```python
rugged_df = pd.read_csv('./data/rugged.csv', sep=';')
rugged_df
```

```python
rugged_df.columns
```

```python
no_nulls_df = rugged_df[['country', 'cont_africa', 'rugged', 'rgdppc_2000']].dropna()
no_nulls_df
```

```python
with pm.Model() as m_7H3_with_seychelles:
    a = pm.Normal('a', 0, 3)
    b_a = pm.Normal('b_a', 0, 1)
    b_r = pm.Normal('b_r', 0, 1)
    b_ar = pm.Normal('b_ar', 0, 1)
    
    mu = a + b_a * no_nulls_df['cont_africa'] + \
        b_r * no_nulls_df['rugged'] + \
        b_ar * no_nulls_df['cont_africa'] * no_nulls_df['rugged']
    sigma = pm.Uniform('sigma', 0,5)
    
    logGDP2000_hat = pm.Normal('logGDP2000_hat', mu, sigma, observed=np.log(no_nulls_df['rgdppc_2000']))
    trace_m_7H3_with_seychelles = pm.sample(2000)
```

```python
no_seychelles_df = no_nulls_df[no_nulls_df['country'] != 'Seychelles']
no_seychelles_df
```

```python
with pm.Model() as m_7H3_without_seychelles:
    a = pm.Normal('a', 0, 3)
    b_a = pm.Normal('b_a', 0, 1)
    b_r = pm.Normal('b_r', 0, 1)
    b_ar = pm.Normal('b_ar', 0, 1)
    
    mu = a + b_a * no_seychelles_df['cont_africa'] + \
        b_r * no_seychelles_df['rugged'] + \
        b_ar * no_seychelles_df['cont_africa'] * no_seychelles_df['rugged']
    sigma = pm.Uniform('sigma', 0,5)
    
    logGDP2000_hat = pm.Normal('logGDP2000_hat', mu, sigma, observed=np.log(no_seychelles_df['rgdppc_2000']))
    trace_m_7H3_without_seychelles = pm.sample(2000)
```

```python
pm.forestplot(trace_m_7H3_with_seychelles)
```

```python
pm.forestplot(trace_m_7H3_without_seychelles)
```

```python
pm.posteriorplot.plot_posterior(trace_m_7H3_with_seychelles)
```

```python
pm.posteriorplot.plot_posterior(trace_m_7H3_without_seychelles)
```

##### Ans

The interaction effect of being in Africa and being rugged has decreased by a little bit. The lower bound of the HPD is now slightly below zero. However, most of the posterior distribution is still above zero, with mean 0.245. 

The beta coefficient for ruggedness averages at -0.180, which is smaller in magnitude than the mean posterior distribution of the beta coefficient of the interaction term, so that means that for countries in Africa, the higher the value of ruggedness, the higher the GDP is.

Finally, conduct a model comparison analysis, using WAIC. fit three models to the data without Seychelles:

$$
\begin{equation}
\begin{aligned}
  \text{Model 1}: y_i &\sim \text{Normal}(\mu_i, \sigma) \\
                  \mu_i &= \alpha + \beta_R R_i \\\\
  \text{Model 2}: y_i &\sim \text{Normal}(\mu_i, \sigma) \\
                  \mu_i &= \alpha + \beta_R R_i + \beta_A A_i \\\\
  \text{Model 3}: y_i &\sim \text{Normal}(\mu_i, \sigma) \\
                  \mu_i &= \alpha + \beta_R R_i + \beta_A A_i + \beta_{AR} A_i R_i \\\\
\end{aligned}
\end{equation}
$$

Use whatever priors you think are sensible. Plot the model-averaged predictions of this model set. Do your inferences differ from those in (b)? Why or why not?
